```python
from pprint import pprint
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import openweather 
import requests
import json
import random
from config import api_key
```

```python
#imported list of all cities available in openweather to avoid getting getting a city not to be found
all_cities = pd.read_json('/Users/edwardkohn/Desktop/Weatherpy/city.list.json')
all_cities.head()
#get list of cities available in openweather
city_id =list(all_cities['id'])
list_of_cities = random.sample(city_id, 550)
```

```python
#url to get city data using the 
url = "http://api.openweathermap.org/data/2.5/weather?units=imperial"

#lists to use
name = []
citiid = []
humidity = []
cloudiness = []
windspeed = []
latitude = []
temperature = []

print('Beginning Data Retrieval') 
print('-----------------------------')

number_record = 0
number_set = 0

#loop to store each city data into lists 
for citi_id in list_of_cities:
    r_url = url + '&appid=' + api_key + '&id=' + str(citi_id)
    response = requests.get(r_url).json()
    
    #to populate lists
    humidity.append(response['main']['humidity'])
    cloudiness.append(response['clouds']['all'])
    windspeed.append(response['wind']['speed'])
    latitude.append(response['coord']['lat'])
    name.append(response['name'])
    citiid.append(response['id'])
    temperature.append(response['main']['temp'])
    
    
    #to print each record
    name_for_print = response['name']
    id_for_print =  response['id']
    
    if record < 50:
        number_record +=1
    
    else:
        number_record = 0
        number_set += 1
        
    print(f'Processing Record {record} of Set {number_set}| {name_for_print}http://api.openweathermap.org/data/2.5/weather?units=Imperial&id={id_for_print}&APPID={api_key}&q={name_for_print}')
    
print('--------------------------')    
print('End Data Retrieval')
```

```python
#construction of df

cities_dict = {'name':name,'citi id':citiid,'humidity':humidity,'cloudiness':cloudiness,'windspeed':windspeed,'latitude':latitude,'temperature':temperature}
cities_df=pd.DataFrame(cities_dict)
cities_df.head()
cities_df.to_csv('Weather_df.csv')
```

# temperature(F) vs Latitude

```python
#temperature vs Latitude
temp_lat_df = cities_df[['temperature','latitude']]
fig,ax = plt.subplots()
ax.scatter(temp_lat_df['latitude'],temp_lat_df['temperature'],marker='.')
ax.set_xlabel('Latitude')
ax.set_ylabel('Temperature(F)')
plt.title('City Latitude vs Temperature(F)')
ax.grid()


fig.savefig('City Latitude vs Average Temperature(F).png')
plt.show()
```

# First Oberservation:


based on the graph above we can see that the tendency is that cities located at lower latitudes tend to have  higher temperatures and cities located at higher latitudes tend to have lower temperatures.

# Humidity (%) vs. Latitude

```python
#Humidity (%) vs. Latitude
hum_lat_df = cities_df[['humidity','latitude']]
fig1,ax1 = plt.subplots()
ax1.scatter(hum_lat_df['latitude'],hum_lat_df['humidity'],marker='.',color='g')
ax1.set_xlabel('Latitude')
ax1.set_ylabel('humidity(%)')
plt.title('City Latitude vs Humidity(%)')
ax1.grid()
fig1.savefig('City Latitude vs Humidity(%).png')
plt.show()
```

# Second Oberservation:


Based on the graph above we can see that most of the cities located from equator up to a latitude of ~50 degrees have the highest humidity while those cities located below the equator tend to have less humidity.

# Cloudiness (%) vs. Latitude

```python
#Cloudiness (%) vs. Latitude
cloud_lat_df = cities_df[['cloudiness','latitude']]
fig2,ax2 = plt.subplots()
ax2.scatter(cloud_lat_df['latitude'],cloud_lat_df['cloudiness'],marker='.',color='r')
ax2.set_xlabel('Latitude')
ax2.set_ylabel('cloudiness(%)')
plt.title('City Latitude vs Cloudiness(%)')
ax2.grid()
fig2.savefig('City Latitude vs Cloudiness(%).png')
plt.show()
```

# Third Observation:

 Based on the sample we have most of the cities where located from latitude from range 40 to 50 degrees. We can determine that at latitude 50 degrees we tend to have more cloudiness.

# Wind Speed (mph) vs. Latitude


```python
#Wind Speed (mph) vs. Latitude
wind_lat_df = cities_df[['windspeed','latitude']]
fig3,ax3 = plt.subplots()
ax3.scatter(wind_lat_df['latitude'],wind_lat_df['windspeed'],marker='.',color='b')
ax3.set_xlabel('Latitude')
ax3.set_ylabel('windspeed(mph)')
plt.title('City Latitude vs Windspeed(MPH)')
ax3.grid()
fig3.savefig('City Latitude vs Windspeed(MPH).png')
plt.show()
```
